```python
df['Order Date'] = pd.to_datetime (df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Monthly Sales Analysis
```

```python
# This code generates a line plot showing the total sales for each month.
Sales_by_month = df.groupby('Order Month')['Sales'].sum().reset_index()
fig =  px.line(Sales_by_month,
                 x ='Order Month',
                 y = 'Sales',
                 title = "Sales by month ")
fig.show() # Added to display the plot
```

```python
df.head()
```

```python
# Sales Analysis by Category
```

```python
Sales_by_category = df.groupby ('Category') ['Sales'].sum().reset_index()
```

```python
Sales_by_category
```

```python
# This code generates a pie chart showing the distribution of sales across different product categories.
fig = px.pie(Sales_by_category,
             values = 'Sales',
             names = 'Category',
             hole = .5,
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition = 'inside', textinfo='percent+label')
fig.update_layout(title_text= "Sales Analysis by Category", title_font=dict(size=24))
fig.show()
```

```python
# Sales Analysis by Sub-Category
```

```python
df.head(1)
```

```python
# This code generates a bar chart showing the total sales for each sub-category.
Sales_by_Sub_Category = df.groupby('Sub-Category')['Sales'].sum().reset_index()
fig = px.bar(Sales_by_Sub_Category,
             x= "Sub-Category",
             y = 'Sales', title = "Sales by Sub Category", color = 'Sub-Category' )
fig.show()
```

```python
# Monthly Profit Analysis
```

```python
df.head(1)
```

```python
# This code generates a line plot showing the total profit for each month.
Profit_by_Month = df.groupby('Order Month')['Profit'].sum().reset_index()
fig =  px.line (Profit_by_Month,
                x ='Order Month' ,
                y ='Profit', title = 'Profit by Month' )
fig.show()
```

```python
# Profit Analysis by Category
```

```python
# This code generates a bar chart showing the total profit for each product category.
Profit_by_Category = df.groupby('Category')['Profit'].sum().reset_index()
fig= px.bar(Profit_by_Category,
            x = 'Category',
            y = 'Profit', title = 'Profit by Category ')
fig.show()
```

```python
# Profit Analysis by Sub-Category
```

```python
# This code generates a line plot showing the total profit for each sub-category.
Profit_by_sub_categoty = df.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.line(Profit_by_sub_categoty,
              x ='Sub-Category',
              y = 'Profit',
              title = "Profit by Sub-Category", )
fig.show()
```

```python
#Sales and Profit Analysis by Customer Segment
```

```python
df.head(1)
```

```python
# This code calculates and plots the total sales and profit for each customer segment.
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment', yaxis_title='Amount')

fig.show()
```

```python
# Analyse sales-to-profit ratio
```

```python
df.head(1)
```

```python
# This code calculates the total profit for each sales amountand gives the ratio of it.
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```
